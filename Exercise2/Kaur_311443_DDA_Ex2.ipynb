```python
size = 4
main = "C:/DDA/20_newsgroups"
directory = os.scandir(main)
list_files = []
for i in directory:
    newpath = os.path.join(i)
    newpath2 = main + '/' + newpath.split("\\")[-1]
    list_files.append(newpath2)

num_files = len(list_files)
step = num_files//size

files_splitted = list_files[: (size - 1)*step]         # Splitting the total number of files
set_files = []

for i in range(size - 1):
    set_files.append(files_splitted[i*step:(i+1)*step])
set_files.append(list_files[(size - 1)*step: ])
print(set_files)
```

```python
set_files = set_files[0]
set_files
```

```python
clean_data = []
for path in set_files:
    sub_directory = os.scandir(path)
    for j in sub_directory:
        files = []
        filepath = os.path.join(j)
        newpath3 = path + '/' + filepath.split("\\")[-1]
        f = open(newpath3, "r")
        file = f.read()
        file = re.sub(r'[^\w\s]', '', file)
        file = re.sub(r'[\d]', '', file)
        file = re.sub(r'\b\w{1}\b', '', file)
        tokens = nltk.word_tokenize(file)
        stop_words = stopwords.words('english')
        for word in tokens:
            if word.lower() not in stop_words:
                files.append(word)
        clean_data.append(files)
```

```python
len(clean_data)
```

```python
with open('pro_data.txt', 'w') as f:
    for doc in clean_data:
        for line in doc:
            f.write(line)
            f.write(' ')
        f.write('\n')
```
