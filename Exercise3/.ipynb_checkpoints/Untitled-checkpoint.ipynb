```python
M, N = data.shape
K = 3

centroids_index = []
for i in range(K):
    centroids_index.append(np.random.randint(0,M))

centroids = np.zeros((K, N))
for i in range(K):
    for j in range(N):
        centroids[i, j] = data[i, j]
```

```python
Distance_matrix = np.zeros((M, K))
for i, ele in enumerate(data):
    for j, center in enumerate(centroids):
        Distance_matrix[i, j] = np.linalg.norm(ele - center)
```

```python
Distance_matrix
```

```python
membership = np.argmin(Distance_matrix, axis=1).reshape(-1, 1)
membership.shape
```
